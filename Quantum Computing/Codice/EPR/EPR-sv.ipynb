```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_state_city
import numpy as np
```

```python
# Define the circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.cx(0,1)

# We wish to save the (non-observable!) full quantum state at this point of the circuit
qc.save_statevector()

# Setting up the simulator
sim = AerSimulator(method="statevector")
qc = transpile(qc, sim)

# Run a simulation
result = sim.run(qc).result()

# Plot the results
sv = result.get_statevector(qc)
plot_state_city(sv, title='EPR states')
```

```python
qc.draw("mpl")
```
