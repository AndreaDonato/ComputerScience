```python
# EPR using Sampler and *exact* simulator (StatevectorSampler)

# More info  https://quantum.cloud.ibm.com/docs/en/guides/primitive-input-output

from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram
from qiskit.transpiler import generate_preset_pass_manager
from qiskit.primitives import StatevectorSampler
from qiskit_ibm_runtime import SamplerV2 as Sampler
import numpy as np
```

```python
# Define the circuit
qc = QuantumCircuit(2)
qc.h(0)      # Hadamard
qc.x(1)      # NOT
qc.cx(0,1)   # Controlled-NOT

# Measure all the qubits
qc.measure_all()

# Draw the circuit
qc.draw("mpl")
```

```python
# Initialize the sampler
sampler = StatevectorSampler()

# Create the ISA circuit for the sampler
pm = generate_preset_pass_manager(optimization_level=1)
isa_circuit = pm.run(qc)
```

```python
# Simulate the circuit
job = sampler.run([isa_circuit], shots=1000)
result = job.result()

# Get results and plot them
pub_result = result[0]
counts = pub_result.data.meas.get_counts()
plot_histogram(counts, title='EPR counts')
```
